```python
results_sph = []
time_results_sph = []
for i in range(len(a)):
    print(i)
    f = Factorizer.Factorizer(a[i], b[i], p[i], method="sph")
    start_time = time.time()
    result = f.factorize()
    end_time = time.time()
    print("x = ", result)
    elapsed_time = end_time - start_time
    print("Elapsed time:", "{:.6f}".format(elapsed_time), "seconds")
    results_sph.append(result)
    time_results_sph.append(elapsed_time)
```

```python
results = []
time_results = []
for i in range(len(a)):
    print(i)
    f = Factorizer.Factorizer(a[i], b[i], p[i])
    start_time = time.time()
    result = f.factorize()
    end_time = time.time()
    print("x = ", result)
    elapsed_time = end_time - start_time
    print("Elapsed time:", "{:.6f}".format(elapsed_time), "seconds")
    results.append(result)
    time_results.append(elapsed_time)
```

```python
for _ in range(26-len(time_results)):
    time_results.append(300)
```

```python
t1_bf = [time_results[i] for i in range(len(time_results)) if i % 2 == 0]
t1_sph = [time_results_sph[i] for i in range(len(time_results_sph)) if i % 2 == 0]

input_sizes = range(3, int(len(time_results)/2) + 3)


plt.plot(input_sizes, t1_bf, label='Brute force')
plt.plot(input_sizes, t1_sph, label='SPH')
plt.xlabel('Input Size')
plt.xticks([i for i in range(3,16)])
plt.ylabel('Time (seconds)')
plt.title('Comparison of Algorithm Execution Time - Type 1')
plt.legend()
plt.savefig('plot1.png')
plt.show()

```
